```python
# List of columns for which you want to calculate percentiles
columns_to_calculate_percentiles = ['%0-5 Points', '%6-10 Points', '%11-15 Points', '%16-20 Points', '%21-25 Points']

# Calculate the 25th, 50th, and 75th percentiles for the specified columns
percentiles_25 = df_2015[columns_to_calculate_percentiles].quantile(0.25)
percentiles_50 = df_2015[columns_to_calculate_percentiles].quantile(0.50)
percentiles_75 = df_2015[columns_to_calculate_percentiles].quantile(0.75)

# Display the percentile values
print("25th Percentiles:\n", percentiles_25)
print("\n50th Percentiles:\n", percentiles_50)
print("\n75th Percentiles:\n", percentiles_75)
```
